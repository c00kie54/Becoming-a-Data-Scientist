## What is SQL? <a class="anchor" id="chapter1"></a>

SQL stands for structured query language.

Used to query, insert, update and modify data - comunicates between you and the database.

It is made up of statments that are descriptives and makes sql easy to understand. 
It is a non-procedural language - this means you cannot write complete aplications with it but can be used to access data.

SQL is used to read/retreive data, write data and update data.

## Models <a class="anchor" id="chapter3"></a>

***Relational Vs Transactional models***

A relational model allows for easy queerying and data manipulation in an easy and logical way.

A transactional model allows for more operational use.



***Data Model building blocks***

***Entities*** - Person, place thing or event - Distinguishable, unique or distinct.

***Attributes*** - characteristics of an entity.

***Relationships*** - Describes associations among entities:
    * One-to-Many
    * Many-to-Many
    * One-to-One

ER diagrams are often used to describe visually the relationships between different tables.

There are different notations to display relationships:
* Chen notation (1:M for one-to-many, M:N for many-to-many, 1:1 for one-to-one)
* Crows foot(Train tracks represent one, crowsfoot represents many)
* UML class (1.1 is one and 1.* is many)

## Getting Data <a class="anchor" id="chapter4"></a>

We use the ***SELECT*** statement to get data from a table.

We say what we want to select and where we want to select it from

```python
## Example - this will display a single columns

SELECT prod_name
FROM Products;
```

```python
## Example - this will display a more columns

SELECT  prod_name
        ,prod_ID 
        ,prod_price
FROM Products;
```

```python
## Example - We can use wild card to get all columns

SELECT *
FROM Products;
```

```python
## Example - We can put a limit so we only view a sample

"""Different Syntaxes are used for different DBMS """

# SQLite
SELECT * 
FROM Products
Limit 5;


# Oracle 
SELECT *
FROM Products
WHERE ROWNUM <=5;

# DB2 
SELECT *
FROM Products 
FETCH FIRST 5 ROWS ONLY;
```
